```python
#Q1.Create a 1D NumPy array of integers from 10 to 50 (inclusive) and extract only the elements that are multiples of 5.
```

```python
result = np.arange(10,51)
result = result[result % 5 ==0]
print(result)
```

```python
#Q2.Generate a 3×3 matrix with values ranging from 2 to 10, then find its transpose.
```

```python
arr1 = np.arange(2,11).reshape(3,3).transpose()
arr1
```

```python
#arr = np.array([[2, 4, 6],
                [1, 3, 5],
                [7, 9, 11]])
#Find the sum of each column and the maximum value of each row.
```

```python
arr = np.array([[2, 4, 6],
                [1, 3, 5],
                [7, 9, 11]])
sum1 = np.sum(arr, axis = 0) #column-wise
print(sum1)
max = np.max(arr, axis = 1) #row-wise 
print(max)
```

```python
#Q4.Create two 3×3 matrices filled with random integers (between 1 and 10) and perform matrix multiplication.
```

```python
matrix1 = np.random.randint(1,11, size=(3,3))
print(matrix1)

matrix2 = np.random.randint(1,11, size= (3,3))
print(matrix2)

matrix_product = np.matmul(matrix1, matrix2)
print(matrix_product)
```

```python
#Q5. Create a 4×4 NumPy array with values from 1 to 16, then replace all elements divisible by 3 with -1.
```

```python
arr2 = np.arange(1,17).reshape(4,4)
arr2
```

```python
arr2[arr2 % 3 == 0] = -1
print(arr2)
```
